```python
#lo primero que haremos será importarnos las librería necessarias para el tratamiento y exploración de datos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
#nos importamos los csv para su vuisualización usando el método pd.read_csv de pandas

df_activity = pd.read_csv("files/customer_flight_activity.csv", index_col = 0)

# hacemos un ".head()" para mostrar las 5 primeras filas del dataframe
df_activity.head()
```

```python

df_loyalty = pd.read_csv("files/customer_loyalty_history.csv", index_col = 0)

# hacemos un ".head()" para mostrar las 5 primeras filas del dataframe
df_loyalty.head()
```

```python
#veamos ahora que dimensiones tienen nuestros dataframes

print(f"El número de filas que tenemos en el dataframe de 'activity' es {df_activity.shape[0]}, y el número de columnas es {df_activity.shape[1]}")
print(f"El número de filas que tenemos en el dataframe de 'loyalty' es {df_loyalty.shape[0]}, y el número de columnas es {df_loyalty.shape[1]}")
```

```python
# veamos ahora una descripción de las columnas de tipo numérico (en el caso de este df son todas)

df_activity.describe().T
```

```python
df_loyalty.describe().T
```

**interpretación**

- `count`: El número de observaciones no nulas en las columnas.

- `mean`: La media de los valores en la columnas.

- `std`: La desviación estándar, que mide la dispersión del conjunto de valores.

- `min`: El mínimo en el conjunto de datos.

- `25%`: El percentil 25, que representa el valor por debajo del cual se encuentra el 25% de los datos.

- `50%`: La mediana o percentil 50, que es el valor que divide el conjunto de datos en dos mitades iguales.

- `75%`: El percentil 75, que representa el valor por debajo del cual se encuentra el 75% de los valores.

- `max`: El valor máximo en el conjunto de datos.


```python
# hagamos lo mismo para las columnas categóricas, para eso tendremos que incluir entre los paréntesis el parámetro include = "object"

df_loyalty.describe(include = "object").T
```

**interpretación**


- `count`: El número de observaciones no nulas en las columnas.

- `unique`: La cantidad de valores únicos en las columnas.

- `top`: El valor más común.

- `freq`: La frecuencia del valor más común.

```python
# el df de actividad no tiene valores nulos
df_activity.isnull().sum()
```

```python
#en este caso, encontramos algunas de las columnas con valores nulos
df_loyalty.isnull().sum()
```

```python
#veamos el porcentaje de nulos de esas columnas 
df_loyalty.isna().sum()/df_loyalty.shape[0]*100
```

```python
# lo convertimos a DataFrame
df_nulos = pd.DataFrame(df_loyalty.isna().sum()/df_loyalty.shape[0]*100, columns = ["%_nulos"])

# filtramos el DataFrame para quedarnos solo con aquellas columnas que tengan nulos
df_nulos[df_nulos["%_nulos"] > 0]
#podemos ver que hay dos columnas que tienen un elevado porcentaje de nulos (casi un 90%). Más adelante veremos cómo lo gestionaremos.
```

```python
#veamos ahora las columnas duplicadas que hay en  nuestros df

print (f'el df de loyalty tiene, {df_loyalty.duplicated().sum()}, filas duplicadas')
```
